```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Create a 3D plot with x, y, and y**2 as x_points, y_points, and z_points respectively
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

figure = plt.figure()

ax = plt.axes(projection="3d")

z_line = y**2
x_line = x
y_line = y

ax.plot3D(x_line, y_line, z_line, 'red')
```
