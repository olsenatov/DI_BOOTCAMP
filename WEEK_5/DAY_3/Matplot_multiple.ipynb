### 2 plots in one figure

```python
figure, axes_set = plt.subplots(2)
```

```python
x = np.linspace(0, 100, 1000)
y1 = x ** 2
y2 = x ** 3

figure, axes_set = plt.subplots(2)
axes_set[0].plot(x, y1, color = "red")
axes_set[1].plot(x, y2, color = "green")
```

```python
x = np.linspace(0, 100, 1000)
y1 = x ** 2
y2 = x ** 3

figure, axes_set = plt.subplots(2)
axes_set[0].plot(x, y1, color = "red")
axes_set[0].annotate('breakout!', xy=(0,0), xytext = (1, 2000), arrowprops={'facecolor' : 'black', 'shrink' : 10})
axes_set[1].plot(x, y2, color = "green")
```

```python
x = np.linspace(0, 100, 1000)
y1 = x ** 2
y2 = x ** 3

figure, axes_set = plt.subplots(2, sharex = True) #sharex - if the x is shared
axes_set[0].plot(x, y1, color = "red", label = "h1n1")
axes_set[0].legend(title = 'Legend', loc = 'lower right', shadow = True)
axes_set[0].annotate('breakout!', xy=(0,0), xytext = (1, 2000), arrowprops={'facecolor' : 'black'})
axes_set[1].plot(x, y2, color = "green", label = "tub")
axes_set[1].legend(title = 'Legend', loc = 'lower right', shadow = True)

plt.tight_layout() #to look clean
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

figure, axes_set = plt.subplots(4)
axes_set[0].plot(x, y)
axes_set[1].plot(x, -y)
axes_set[2].plot(x, y**2)
axes_set[3].plot(-y**2, x)

plt.tight_layout()
```

```python
figure, axes_set = plt.subplots(2, sharex = True) #sharex - if the x is shared
axes_set[0].plot(x, y1, color = "red", label = "h1n1")
axes_set[0].legend(title = 'Legend', loc = 'lower right', shadow = True)
axes_set[0].annotate('max!', xy=(0,0), xytext = (1, 2000), arrowprops={'facecolor' : 'black'})
axes_set[0].annotate('min', xy=(0,0), xytext = (1, 2000), arrowprops={'facecolor' : 'black'})
axes_set[1].plot(x, y2, color = "green", label = "tub")
axes_set[1].legend(title = 'Legend', loc = 'lower right', shadow = True)
```
