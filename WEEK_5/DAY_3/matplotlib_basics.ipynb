```python
### Histogram - collecting the data into bins based on value
#example - distribution of how many people in the exact height range
```

```python
# data = np.random.rand(1000)
data = [1,1,1,1,1,5,5,5,5,5,5,5,5,9,9,9,9,9,9]
figure, axes = plt.subplots()
axes.hist(data, bins = 3)
```

### Stylize

```python
figure, axes = plt.subplots()

##color, marker, labels
axes.set(title = "Line plot", xlabel = "X", ylabel = "Y")
axes.plot([1,2,3,4], [2,4,5,7], color = "red", linestyle = "--", marker = "o", mfc = "green", mec = "black")
```

```python
## the legend
figure, axes = plt.subplots()

##color, marker, labels
axes.set(title = "Line plot", xlabel = "X", ylabel = "Y")
axes.plot([1,2,3,4], [2,4,5,7], color = "red", linestyle = "--", marker = "o", mfc = "green", mec = "black", label = "Yossi")
axes.plot([3,4,5,6], [8,1,4,5], color = "blue", linestyle = "--", marker = "o", mfc = "red", mec = "black", label = "Ziv")
axes.legend()

#for VisualStudioCode plt.show()
```

```python
#saving
figure, axes = plt.subplots()

axes.set(title = "Line plot", xlabel = "X", ylabel = "Y")
axes.plot([1,2,3,4], [2,4,5,7], color = "red", linestyle = "--", marker = "o", mfc = "green", mec = "black", label = "Yossi")
axes.plot([3,4,5,6], [8,1,4,5], color = "blue", linestyle = "--", marker = "o", mfc = "red", mec = "black", label = "Ziv")
axes.legend()

figure.savefig("my_plot.png")
```

```python
x2 = np.linspace(0, 10, 20)
y3 = x2 ** 2
y4 = x2 ** 1.5

figure, axes = plt.subplots()
axes.set(title = "My style plot", xlabel = "X", ylabel = "Y")
axes.plot(x2, y3, color = "red", linestyle = "dashed", marker = "o", label = "y=x^2")

axes.plot(x2, y4, color = "blue", linestyle = "dotted", marker = "s", label = "y=x^1,5")
axes.legend()
```
