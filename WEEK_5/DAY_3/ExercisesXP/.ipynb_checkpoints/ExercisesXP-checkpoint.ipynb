```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

figure, axes = plt.subplots()
axes.set(title = "y = cos(x)")
axes.scatter(x, y2, color = 'blue')
```

```python
#8
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

figure, axes_set = plt.subplots(2, sharex = True)
axes_set[0].plot(x, y1, color = 'blue')
axes_set[1].plot(x, y2, color = 'blue')
```

```python
#9
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

figure, axes_set = plt.subplots(2, sharex = True)
axes_set[0].plot(x, y1, color = 'blue', label = 'y = sin(x)')
axes_set[0].legend(loc = 'upper left')
axes_set[1].plot(x, y2, color = 'blue', label = 'y = cos(x)')
axes_set[1].legend( loc = 'center right')
```
