```python
starbucks['Caffeine (mg)'] = pd.to_numeric(starbucks['Caffeine (mg)'], errors='coerce')
```

```python
starbucks['caffeine_to_calories'] = pd.to_numeric(starbucks['Caffeine (mg)']) / starbucks['Calories']
```

```python
starbucks['caffeine_to_calories']
```

```python
#12
averages = starbucks.groupby('Beverage_category')['caffeine_to_calories'].mean().to_frame()
averages
averages.sort_values(by='caffeine_to_calories', ascending=False).head(1)
```

```python
#13
import matplotlib.pyplot as plt
```

```python
for_plot = starbucks.groupby('Beverage_category')['caffeine_to_calories'].mean().to_frame()
for_plot
```

```python
cat = for_plot.index
val = for_plot['caffeine_to_calories']
figure = plt.figure(figsize = (10,5))
plt.barh(cat, val)
```

```python
#14
coffein = starbucks[' Sugars (g)']
sug = starbucks['Caffeine (mg)']
cof_to_sug = pd.DataFrame({
    'coffein' : coffein, 'sugar' : sug})
```

```python
cof_to_sug
```

```python
figure, axes = plt.subplots()
axes.plot(coffein, label = "coffein")
axes.plot(sug, color="red", label = "sugar")
axes.legend()
```
