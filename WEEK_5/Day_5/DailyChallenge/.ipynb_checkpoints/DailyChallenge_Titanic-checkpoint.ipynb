```python
# the ratio of male to female passengers
gender_ratio = titanic_df['Sex'].value_counts()

# pie chart
plt.figure(figsize=(6, 6))
plt.pie(gender_ratio, labels=gender_ratio.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'pink'])
plt.title('Ratio of Male to Female Passengers')
plt.axis('equal') 
plt.show()
```

```python
gender_ratio
```

```python
#age distribution
plt.figure(figsize=(8, 6))
plt.hist(titanic_df['Age'].dropna(), bins=20, color='orange', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Passenger Ages')
plt.show()
```

```python
#survival count
survival_count = titanic_df['Survived'].value_counts()

plt.figure(figsize=(6, 6))
plt.bar(survival_count.index, survival_count.values, color=['lightgreen', 'red'])
plt.xticks(survival_count.index, ['Survived', 'Did Not Survive'])
plt.xlabel('Survival')
plt.ylabel('Number of Passengers')
plt.title('Survival Counts')
plt.show()
```

```python
#survival count split by gender
gender_survival_count = titanic_df.groupby(['Sex', 'Survived']).size().unstack()

plt.figure(figsize=(8, 6))
gender_survival_count.plot(kind='bar', stacked=True, color=['red', 'green'])
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.title('Survival Counts by Gender')
plt.legend(['Did Not Survive', 'Survived'])
plt.show()
```

```python
gender_survival_count
```

```python
#correlation between fare and age
plt.figure(figsize=(8, 6))
plt.scatter(titanic_df['Age'], titanic_df['Fare'], color='orange', alpha=0.4)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Correlation between Fare and Age')
plt.show()
```

```python
# Replace missing values in 'Age' column with the median age
median_age = titanic_df['Age'].median()
median_age
```

```python
titanic_df['Age'].fillna(median_age, inplace=True)
```

```python
titanic_df['Age']
```

```python
#normalize fare column
titanic_fare_norm = (titanic_df['Fare']- titanic_df['Fare'].min())/(titanic_df['Fare'].max()- titanic_df['Fare'].min())
```

```python
titanic_fare_norm
```

```python
#give numerical representation to sex column (0 - men, 1 - female)
titanic_df['Sex'] = titanic_df['Sex'].apply(lambda x: 1 if x == 'female' else 0)
```

```python
titanic_df
```

```python
#survival rate per gender (0 - men, 1 - female)

titanic_df.groupby('Sex')['Survived'].mean() * 100 (#get the real percentage)
```

```python
#survival rate per passanger classes
titanic_df.groupby('Pclass')['Survived'].mean() * 100
```
