## Part 2

```python
new_properties = boston_housing[boston_housing['age'] <= 35.0]
middle_properties = boston_housing[(boston_housing['age'] > 35.0) & (boston_housing['age'] <= 70.0)]
old_properties = boston_housing[boston_housing['age'] > 70.0]
plt.figure(figsize=(10, 6))

plt.hist(new_properties['medv'], bins=20, alpha=1.0, label='New Properties')
plt.hist(middle_properties['medv'], bins=20, alpha=0.8, label='Middle Properties')
plt.hist(old_properties['medv'], bins=20, alpha=0.4, label='Old Properties')

plt.xlabel('Median Value (MEDV)')
plt.ylabel('Frequency')
plt.title('Property Age vs. Median Value')
plt.legend()
plt.show()
```

## Part 3

```python
boston_housing['crim'] = pd.cut(boston_housing['crim'], bins=[0, 1, 5, 10, 20, max(boston_housing['crim'])],
                        labels=['0-1', '1-5', '5-10', '10-20', '20+'])
```

```python
boston_housing['crim']
```

```python
avg_medv_per_crim_bin = boston_housing.groupby('crim')['medv'].mean()
```

```python
avg_medv_per_crim_bin
```

```python

plt.figure(figsize=(10, 6))
avg_medv_per_crim_bin.plot(kind='line', marker='o', color='b')

plt.xlabel('Binned Crime Rate')
plt.ylabel('Average MEDV')
plt.title('Impact of Neighborhood Crime Rate on Prices')
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.show()
```
